```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot graph for problem (a)
runs = np.arange(1, 11)

# One worker
run_times_one_worker = [4.9, 7.04, 9.16, 11.61, 13.59, 15.64, 17.97, 20.04, 22.60, 24.71]
run_times_one_worker = np.array(run_times_one_worker)

fig, ax = plt.subplots()

ax.scatter(runs, run_times_one_worker)
ax.set_ylim(0, 26)
ax.set_xlabel("Run Number")
ax.set_ylabel("Runtime (seconds)")
ax.set_title("Runtime vs Run Number One Worker")

plt.show()
```

```python
# Plot graph for problem (b)
run_times_ten_workers = [3.83, 4.73, 5.63, 6.59, 7.51, 8.44, 9.41, 10.19, 11.10, 12.02]
run_times_ten_workers = np.array(run_times_ten_workers)

fig, ax = plt.subplots()

ax.scatter(runs, run_times_ten_workers)
ax.set_ylim(0, 26)
ax.set_xlabel("Run Number")
ax.set_ylabel("Runtime")
ax.set_title("Runtime vs Run Number 10 Workers")

plt.show()
```

```python
one_worker_avg_runtime = np.mean(run_times_one_worker)
ten_workers_avg_runtime = np.mean(run_times_ten_workers)

speedup = one_worker_avg_runtime / ten_workers_avg_runtime
print("Speedup:", speedup)
```

Taking the average runtimes for one worker and ten workers, I calculate the speedup to be 1.85. While this is a large speedup, it doesn't seem too big considering we increased the amount of workers by 10.

```python
run_times_2 = [3.5, 4.27, 5.0, 5.8, 6.54, 7.3, 8.13, 8.78, 9.75, 10.6]
run_times_2 = np.array(run_times_2)

fig, ax = plt.subplots()

ax.scatter(runs, run_times_one_worker, label="One Worker")
ax.scatter(runs, run_times_ten_workers, label="Ten Workers 1")
ax.scatter(runs, run_times_2, label="Ten Workers 2")
ax.set_ylim(0, 26)
ax.set_xlabel("Run Number")
ax.set_ylabel("Runtime (seconds)")
ax.set_title("Runtime vs Run Number One Worker")

ax.legend()

plt.show()

```

```python
avg_runtime_2 = np.mean(run_times_2)

speedup = ten_workers_avg_runtime / avg_runtime_2
print("Speedup:", speedup)
```

There is a speedup of 1.14 between part 2 and 3, even though the operations performed were the same. This means that the only thing that changed was the order of the operations, which can effect the performance of the system because the CPU is not a stateless resource, meaning the order of the operations can effect the performance. This is because if similar operations are performed sequentially there will be greater cache hits, and thus greater performance. In order to increase the performance of the system, one could use a bounded First Ready, First Come First Served queueing policy, where requests are re-ordered according to the current operation being executed. If a request is the same as the current operation being executed, it will be prioritized, otherwise it will be first come first serve. In order to prevent starvation there will be a  bound to how many requests can get priority in a row.
